```python
import seaborn as sns 
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips 
```

```python
#scatter plot 
sns.scatterplot(data=tips,x="total_bill",y='tip')
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",hue="sex",style="time",size="size")
```

```python
gap = px.data.gapminder()
gap
```

```python
nepal_data = gap[gap["country"] == 'Nepal']
```

```python
sns.relplot(kind="line", data=nepal_data,x='year',y='lifeExp')
```

```python
temp_df = gap[gap["country"].isin(['India','Nepal','China','Germany','Brazil'])]
```

```python
temp_df 
```

```python
sns.relplot(kind="line",data=temp_df,x="year",y='lifeExp',hue="country",style="continent")
```

```python
#facetPlot
sns.relplot(data=tips,x='total_bill',y='tip',kind="scatter",row="sex",col='smoker')
```

```python
sns.relplot(data=gap,x="lifeExp",y="gdpPercap",kind="scatter",col="year",col_wrap=2)
```

```python
#distribution plot 
#figure-level-plot --> displot
#axeslevelplot --> histplot --> kdeplot --> rugplot 
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
 sns.displot(data=tips,x="total_bill" ,kind="hist")
```

```python
 sns.displot(data=tips,x="total_bill" ,kind="hist",bins=20)
```

```python
#countPlot
sns.displot(data=tips,x="day" ,kind="hist")
```

```python
 sns.displot(data=tips,x="tip" ,kind="hist",hue="sex")
```

```python
 sns.displot(data=tips,x="tip" ,kind="hist",hue="sex",element="step")
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
 sns.displot(data=titanic,x="age" ,kind="hist",element="step",hue="sex")
```

```python
#kdePlot --> KERNEL DENSITY ESTIMATION  --> produces continuous density estimates 
print(tips)
sns.kdeplot(data=tips,x="total_bill")
```

```python
sns.displot(data=tips,x="total_bill",kind="kde",hue="sex",fill=True)
```
