```python
import seaborn as sns 
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips 
```

```python
#scatter plot 
sns.scatterplot(data=tips,x="total_bill",y='tip')
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",hue="sex",style="time",size="size")
```

```python
gap = px.data.gapminder()
gap
```

```python
nepal_data = gap[gap["country"] == 'Nepal']
```

```python
sns.relplot(kind="line", data=nepal_data,x='year',y='lifeExp')
```

```python
temp_df = gap[gap["country"].isin(['India','Nepal','China','Germany','Brazil'])]
```

```python
temp_df 
```

```python
sns.relplot(kind="line",data=temp_df,x="year",y='lifeExp',hue="country",style="continent")
```

```python
#facetPlot
sns.relplot(data=tips,x='total_bill',y='tip',kind="scatter",row="sex",col='smoker')
```

```python
#kdePlot --> KERNEL DENSITY ESTIMATION  --> produces continuous density estimates 
print(tips)
sns.kdeplot(data=tips,x="total_bill")
```

```python
sns.displot(data=tips,x="total_bill",kind="kde",hue="sex",fill=True)
```

```python
#RUGPLOT 
sns.kdeplot(data=tips,x="total_bill")
sns.rugplot(data=tips,x="total_bill")

```

```python
#BIVARIATE HISTOGRAM 
# sns.histplot(data=tips,x="total_bill",y="tip")
sns.displot(data=tips,x="total_bill",y="tip",kind="hist")
```

```python
#BIVARIATE KDEPLOT 
sns.displot(data=tips,x="total_bill",y="tip",kind="kde")
```

```python
#HEATMAP 
country_temp = gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
#axes level function 
plt.figure(figsize=(20,10))
sns.heatmap(country_temp)
```

```python
european_countries = gap[gap["continent"] == "Europe"].pivot(index="country",columns="year",values="lifeExp")
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(european_countries,annot=True,linewidth=0.5,cmap="summer")
```
