```python
#PAIRPLOT
sns.pairplot(iris,hue="species")
```

```python
#PAIRGRID 
```

```python
g = sns.PairGrid(data=iris,hue="species")
g.map(sns.scatterplot)
```

```python
ge = sns.PairGrid(data=iris,hue="species")
ge.map_diag(sns.boxplot)
ge.map_offdiag(sns.violinplot)
```

```python
gee = sns.PairGrid(data=iris,hue="species",vars=["sepal_width","petal_width"])
gee.map_diag(sns.histplot)
gee.map_upper(sns.violinplot)
gee.map_lower(sns.scatterplot)
```

```python
#JOINPLOT AND JOINTGRID 

sns.jointplot(data=tips,x="total_bill",y="tip",kind="reg")
```

```python
j = sns.JointGrid(data=tips,x="total_bill",y="tip")
j.plot(sns.kdeplot,sns.violinplot)
```
